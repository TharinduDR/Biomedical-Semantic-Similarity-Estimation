```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

## some config values 
embed_size = 300 # how big is each word vector
max_features = None # how many unique words to use (i.e num rows in embedding vector)
maxlen = 26 # max number of words in a question to use #99.99%

## fill up the missing values
X_left = sick_aug_normalized["sent_1"].fillna("_na_").values
X_right = sick_aug_normalized["sent_2"].fillna("_na_").values

X_test_left = sick_test_normalized["sent_1"].fillna("_na_").values
X_test_right = sick_test_normalized["sent_2"].fillna("_na_").values


## Tokenize the sentences
tokenizer = Tokenizer(num_words=max_features, filters='')
tokenizer.fit_on_texts(list(X_right) + list(X_left))

X_test_left = tokenizer.texts_to_sequences(X_test_left)
X_test_right = tokenizer.texts_to_sequences(X_test_right)

X_test_left = pad_sequences(X_test_left, maxlen=maxlen)
X_test_right = pad_sequences(X_test_right, maxlen=maxlen)
```

```python
from embeddings.load_embeddings import load_word2vec

word2vec = load_word2vec("/data/word2vec/GoogleNews-vectors-negative300.bin.gz")
```

```python
import numpy as np

word_index = tokenizer.word_index
max_features = len(word_index)+1

def load_embedding_matrix(word_index):
    
    all_embs = np.stack(word2vec.vectors)
    emb_mean,emb_std = all_embs.mean(), all_embs.std()
    embed_size = all_embs.shape[1]

    embedding_matrix = np.random.normal(emb_mean, emb_std, (max_features, embed_size))
    for word, i in word_index.items():
        if i >= max_features: continue
        if word in word2vec.vocab : embedding_matrix[i] = word2vec.word_vec(word)
            
    return embedding_matrix 
```

```python
embedding_matrix = load_embedding_matrix(word_index)
```

```python
import keras.backend as K


def exponent_neg_manhattan_distance(left, right):
    """ Helper function for the similarity estimate of the LSTMs outputs"""
    return K.exp(-K.sum(K.abs(left - right), axis=1, keepdims=True))
```

```python
from keras import Input, Model
from keras.layers import Embedding, GRU, Lambda
from keras.optimizers import Adadelta

def model():
    # The visible layer
    K.clear_session()
    
    left_input = Input(shape=(maxlen,), dtype='int32')
    right_input = Input(shape=(maxlen,), dtype='int32')

    embedding_layer = Embedding(len(embedding_matrix), embed_size, weights=[embedding_matrix], input_length=maxlen,
                                trainable=False)

    # Embedded version of the inputs
    encoded_left = embedding_layer(left_input)
    encoded_right = embedding_layer(right_input)

    # Since this is a siamese network, both sides share the same LSTM
    shared_gru = GRU(50, name='gru')
    # shared_gru = GRU(n_hidden, name='gru', return_sequences=True,
    #                  kernel_initializer=glorot_normal(seed=12300),
    #                  recurrent_initializer=orthogonal(gain=1.0, seed=10000))

    left_output = shared_gru(encoded_left)
    right_output = shared_gru(encoded_right)

    # Calculates the distance as defined by the MaLSTM model
    magru_distance = Lambda(function=lambda x: exponent_neg_manhattan_distance(x[0], x[1]),
                            output_shape=lambda x: (x[0][0], 1))([left_output, right_output])

    # Pack it all up into a model
    magru = Model([left_input, right_input], [magru_distance])

    optimizer = Adadelta(clipnorm=1.25)

    magru.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    
    return magru
    
```

```python
def f1_smart(y_true, y_pred):
    args = np.argsort(y_pred)
    tp = y_true.sum()
    fs = (tp - np.cumsum(y_true[args[:-1]])) / np.arange(y_true.shape[0] + tp - 1, tp, -1)
    res_idx = np.argmax(fs)
    return 2 * fs[res_idx], (y_pred[args[res_idx]] + y_pred[args[res_idx + 1]]) / 2
```

```python
from sklearn.model_selection import KFold
from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau

X = sick_aug_normalized[['sent_1','sent_2']]
Y = sick_aug_normalized['sim']

bestscore = []
y_test = np.zeros((X_test_left.shape[0], ))

kf = KFold(n_splits=5)
kf.get_n_splits(X)
print (kf)

model = model()

for i, (train_index, valid_index) in enumerate(kf.split(X)):
    X_train, X_val, Y_train, Y_val = X.iloc[train_index], X.iloc[valid_index], Y.iloc[train_index], Y.iloc[valid_index]
    
    X_left = X_train["sent_1"].fillna("_na_").values
    X_right = X_train["sent_2"].fillna("_na_").values
    Y_train = Y_train.values
    
    X_left = tokenizer.texts_to_sequences(X_left)
    X_right = tokenizer.texts_to_sequences(X_right)
    
    X_left = pad_sequences(X_left, maxlen=maxlen)
    X_right = pad_sequences(X_right, maxlen=maxlen)
        
    X_val_left = X_val["sent_1"].fillna("_na_").values
    X_val_right = X_val["sent_2"].fillna("_na_").values
    Y_val = Y_val.values
    
    X_val_left = tokenizer.texts_to_sequences(X_val_left)
    X_val_right = tokenizer.texts_to_sequences(X_val_right)
    
    X_val_left = pad_sequences(X_val_left, maxlen=maxlen)
    X_val_right = pad_sequences(X_val_right, maxlen=maxlen)

           
    filepath="weights_best.h5"
    checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=2, save_best_only=True, mode='min')
    
    callbacks = [checkpoint]
    
    if i == 0:print(model.summary()) 
        
    model.fit([X_left, X_right], Y_train, batch_size=512, epochs=100, validation_data=([X_val_left, X_val_right], Y_val), verbose=2, callbacks=callbacks, 
             )
    model.load_weights(filepath)
    
    y_pred = model.predict([X_val_left, X_val_right], batch_size=1024, verbose=2)
    y_test += np.squeeze(model.predict([X_test_left, X_test_right], batch_size=1024, verbose=2))/5
    f1, threshold = f1_smart(np.squeeze(Y_val), np.squeeze(y_pred))
    print('Optimal F1: {:.4f} at threshold: {:.4f}'.format(f1, threshold))
    bestscore.append(threshold)
```

```python
import scipy
from math import sqrt
from sklearn.metrics import mean_squared_error

pearson_correlation = scipy.stats.pearsonr(y_test, sick_test_normalized['sim'])[0]
spearman_correlation = scipy.stats.spearmanr(y_test, sick_test_normalized['sim'])[0]
rmse = sqrt(mean_squared_error(y_test, sick_test_normalized['sim']))
textstr = 'RMSE=%.3f\n$Pearson Correlation=%.3f$\n$Spearman Correlation=%.3f$' % (
        rmse, pearson_correlation, spearman_correlation)
```
